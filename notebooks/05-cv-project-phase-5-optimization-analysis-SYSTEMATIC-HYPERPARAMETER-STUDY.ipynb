```python
import shutil
import os

# Define source and destination paths
repo_src = '/kaggle/input/pipnet-phase5-stable'
repo_dst = '/kaggle/working/PIPNet'

# Remove the existing directory to prevent errors during the copy process
if os.path.exists(repo_dst):
    shutil.rmtree(repo_dst)

# Copy the repository to the working directory
shutil.copytree(repo_src, repo_dst)

print(f"Repository ready at {repo_dst}")
```

```python
# =================================================================
# VERSION 4: PHASE 5 - STEP 2 SYSTEMATIC HYPERPARAMETER STUDY
# =================================================================
# Objective:
# Analyze the impact of 4 critical hyperparameters.
# All runs use 30 Epochs for consistency.
# =================================================================

# Change directory to source code
%cd /kaggle/working/PIPNet

# Define the base command for consistency
base_cmd = "python main.py --dataset COVID19 --batch_size 32 --batch_size_pretrain 32 --epochs_pretrain 10 --epochs 30 --freeze_epochs 10 --lr 0.05 --lr_net 0.0005 --lr_block 0.0005 --weight_decay 0.0 --image_size 224 --seed 1 --weighted_loss --gpu_ids 0 --decay_start_epoch 10 --w_align 5.0 --w_tanh 2.0 --visualize"

print(">>> Starting Systematic Study...")

# ---------------------------------------------------------
# 0. REFERENCE BASELINE (Standard Control Group)
# ---------------------------------------------------------
# Config: ResNet18 | 128 Feats | Decay 0.0008 | W_Class 2.0
# We run this first so it's guaranteed to exist for comparisons.
print("\n--- Running 0: Reference Baseline (Standard) ---")
!{base_cmd} --net resnet18 --decay_rate 0.0008 --num_features 128 --w_class 2.0 --log_dir /kaggle/working/experiment_feasible_optimized

# ---------------------------------------------------------
# A. DECAY RATE STUDY
# ---------------------------------------------------------
# Low (0.0005) vs High (0.001) vs Ref (0.0008)
print("\n--- Running A1: Decay Low (0.0005) ---")
!{base_cmd} --net resnet18 --decay_rate 0.0005 --num_features 128 --w_class 2.0 --log_dir /kaggle/working/exp_decay_low

print("\n--- Running A2: Decay High (0.001) ---")
!{base_cmd} --net resnet18 --decay_rate 0.001 --num_features 128 --w_class 2.0 --log_dir /kaggle/working/exp_decay_high

# ---------------------------------------------------------
# B. NUM FEATURES STUDY (Capacity)
# ---------------------------------------------------------
# Low (64) vs High (256) vs Ref (128)
print("\n--- Running B1: Features Low (64) ---")
!{base_cmd} --net resnet18 --decay_rate 0.0008 --num_features 64 --w_class 2.0 --log_dir /kaggle/working/exp_feat_low

print("\n--- Running B2: Features High (256) ---")
!{base_cmd} --net resnet18 --decay_rate 0.0008 --num_features 256 --w_class 2.0 --log_dir /kaggle/working/exp_feat_high

# ---------------------------------------------------------
# C. CLASS WEIGHT STUDY (Interpretability Trade-off)
# ---------------------------------------------------------
# Low (1.0) vs High (5.0) vs Ref (2.0)
print("\n--- Running C1: Class Weight Low (1.0) ---")
!{base_cmd} --net resnet18 --decay_rate 0.0008 --num_features 128 --w_class 1.0 --log_dir /kaggle/working/exp_wclass_low

print("\n--- Running C2: Class Weight High (5.0) ---")
!{base_cmd} --net resnet18 --decay_rate 0.0008 --num_features 128 --w_class 5.0 --log_dir /kaggle/working/exp_wclass_high

# ---------------------------------------------------------
# D. BACKBONE ARCHITECTURE STUDY
# ---------------------------------------------------------
# ResNet34 vs ResNet50 vs Ref (ResNet18)
print("\n--- Running D1: Backbone ResNet34 ---")
!{base_cmd} --net resnet34 --decay_rate 0.0008 --num_features 128 --w_class 2.0 --log_dir /kaggle/working/exp_net_resnet34

print("\n--- Running D2: Backbone ResNet50 ---")
# Using Batch Size 32 ensures this runs safely without OOM
!{base_cmd} --net resnet50 --decay_rate 0.0008 --num_features 128 --w_class 2.0 --log_dir /kaggle/working/exp_net_resnet50

print("\n>>> Systematic Study Complete.")
```
